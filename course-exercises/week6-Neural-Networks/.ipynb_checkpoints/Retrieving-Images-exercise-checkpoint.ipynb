# Task 2: Create category-specific image retrieval models

## Split the training data into 4 different SFrame data structures
### Each will contain data for one of the four categories image categories

```python
training_data_cat = image_data.filter_by('cat', 'label')
training_data_dog = image_data.filter_by('dog', 'label')
training_data_automobile = image_data.filter_by('automobile', 'label')
training_data_bird = image_data.filter_by('bird', 'label')
```

# Create a nearest neighbor model using deep features (one such model for each category)

```python
cat_model = turicreate.nearest_neighbors.create(training_data_cat,
                                               features = ['deep_features'],
                                               label = 'id')
dog_model = turicreate.nearest_neighbors.create(training_data_dog,
                                               features = ['deep_features'],
                                               label = 'id')
automobile_model = turicreate.nearest_neighbors.create(training_data_automobile,
                                               features = ['deep_features'],
                                               label = 'id')
bird_model = turicreate.nearest_neighbors.create(training_data_bird,
                                               features = ['deep_features'],
                                               label = 'id')
```

# What is the nearest cat-labeled image in the training data to the cat image that is the first image in the test data ?

```python
cat = image_test[0:1]
```

```python
cat['image'].explore()
```

```python
# This function will match the 'id' in image_train with 'reference_lebel' in query_result
def get_images_from_ids(query_result):
    return image_data.filter_by(query_result['reference_label'],'id')
```

```python
cat_neighbors = get_images_from_ids(cat_model.query(cat))
```

```python
cat_neighbors
```

```python

cat_neighbors['image'].explore()
```

# Task 4: Compute nearest neighbors accuracy
### In this task, you will measure the accuracy of a 1-nearest-neighbor classifier, i.e., predict the output as the label of the nearest neighbor in the training data

```python
image_test_cat = image_test.filter_by('cat', 'label')
image_test_dog = image_test.filter_by('dog', 'label')
image_test_automobile = image_test.filter_by('automobile', 'label')
image_test_bird = image_test.filter_by('bird', 'label')
```

## Next you'll find nearest neighbors in the training set for each part of the test set

### Find the closest neighbor to the dog test data using each of the trained models

```python
dog_cat_neighbors = cat_model.query(image_test_dog, k=1)
dog_dog_neighbors = dog_model.query(image_test_dog, k=1)
dog_automobile_neighbors = automobile_model.query(image_test_dog, k=1)
dog_bird_neighbors = bird_model.query(image_test_dog, k=1)
```

### The distance column in dog_cat_neighbors contains the distance between each dog-labeled image in the test set and its nearest cat-labeled image in the training set

## Create an SFrame with the distances from dog test examples to the respective nearest neighbors in each class in the training data

```python
dog_distances = turicreate.SFrame({'dog-automobile': dog_automobile_neighbors['distance'],
                              'dog-bird': dog_bird_neighbors['distance'],
                              'dog-cat': dog_cat_neighbors['distance'],
                              'dog-dog': dog_dog_neighbors['distance']
                             })
```

```python
dog_distances
```

## Computing the number of correct predictions using 1-nearest neighbors for the dog class

### Create a function which returns 1 if the value for row[dog-dog] is lower than that of the other columns, and 0 otherwise. That is, returns 1 if this row is correctly classified by 1-nearest neighbors, and 0 otherwise.

```python
def is_dog_correct(row):  
    if row['dog-dog'] <= min(row.values()):     
        return 1    
    else:        
        return 0
```

## Computing the number of correct predictions for dog

```python
dog_distances.apply(is_dog_correct).sum()
```

###  Accuracy of predicting dog in the test data: Using the work you did in this question, what is the accuracy of the 1-nearest neighbor classifier at classifying dog images from the test set?

```python
dog_distances.apply(is_dog_correct).sum()/float(len(dog_distances))
```
